```python
#Array Slicing: Accessing subarrays using :

#notation: x[start:stop:step]. Without specifying, it will default to start=0, stop = size of dimension, step =1.

#one dimensional arrays:
x = np.arange(10)
x

x[:5] #first five elements

x[5:] #elements after index 5
x[4:7] #elements from index 4 to 6
x[::2] #every other element
x[1::2] #every other element, starting at index 1

#we can use negative step to reverse start and stop. If we wanted to reverse an array:
x[::-1] #all elements, reversed
x[5::-2] #reversed every other from index 5



```

```python
#accessing array rows and columns:
#to do this, combine both indexing and slicing, sing an empty slice marked by a single conon (:):
print(x2[:,0]) #first column of x2 (because row is empty)
print(x2[0,:]) #first row of x2 (because column is empty)

#but for row access, the empty slice can be omitted for a more compact syntax:
print(x2[0]) #basicaly is equivalent to x2[0,:]
```

```python
#sub-arrays as no-copy view, array slices return views rather than copies of array data (numpy array slicing differs from python list slicing in this way). In lists, slices will be copies. 

#ex:

print(x2)

#exctract 2x2 array from this:

x2_sub=x2[:2, :2]
print(x2_sub)
```

```python
#modifying the slice modifies the original:
x2_sub[0,0]=99
print(x2_sub)
```

Rules of Broadcasting:
Rule 1: If the two arrays differ in their number of dimensions, the shape of the one with fewer dimensions is padded with ones on its leading (left) side.
Rule 2: if the shape of the 2 arrays does not match in any dimension, the array with shape equal to 1 in that dimension is stretched to match the other shape
Rule 3: if any dimension the sizes disagree and neither is equal to 1, an error is raised.

Using bitwise operators:

print("Number days without rain:  ", np.sum(rainfall_mm == 0))
print("Number days with rain:     ", np.sum(rainfall_mm != 0))
print("Days with more than 10 mm: ", np.sum(rainfall_mm > 10))
print("Rainy days with < 5 mm:    ", np.sum((rainfall_mm > 0) &
                                            (rainfall_mm < 5)))


```python
#when you want to pull only the subset of the array that meets some criteria, that is called a boolean mask\

# construct a mask of all rainy days
rainy = (rainfall_mm > 0)

# construct a mask of all summer days (June 21st is the 172nd day)
days = np.arange(365)
summer = (days > 172) & (days < 262)

print("Median precip on rainy days in 2015 (mm):   ",
      np.median(rainfall_mm[rainy]))
print("Median precip on summer days in 2015 (mm):  ",
      np.median(rainfall_mm[summer]))
print("Maximum precip on summer days in 2015 (mm): ",
      np.max(rainfall_mm[summer]))
print("Median precip on non-summer rainy days (mm):",
      np.median(rainfall_mm[rainy & ~summer]))



#also ex:
#array = x
x(x<5)

#only selects values from the array that are less than 5
```

```python
#STRUCTURED DATA:NUMPYS STRUCTURED ARRAYS
#imagine we have several categories of data on a number of people - we'd like to store these values for use in Python program
#impossible to use with 3 seperate arrays:

name = ['Alice','Bob','Cathy','Doug']
age = [25,45,37,19]
weight = [55.0,85.5,68.0,61.5]
```

```python
#previously:
x = np.zeros(4,dtype=int)

#using structured array:
data = np.zeros(4, dtype={'names':('name','age','weight'),
                          'formats':('U10','i4','f4')   })
print(data.dtype)
```

```python
data['name'] = name
data['age'] = age
data['weight'] = weight
print(data)
```

```python
#now we can refer to values either by index of name:
data['name']
```

```python
#get first row of data
data[0]
```

```python
#get the name from the last row:
data[-1]['name']
```

```python
#using boolean masking, we can even do some more sophisticated operations, like filtering on age:
data[data['age'] < 30]['name']

#get names where age is under 30 ^
```

```python
#exploring structured array creation:
np.dtype({'names':('name', 'age', 'weight'),
          'formats':('U10', 'i4', 'f8')})
```
