```python
good_reviews=train_df[train_df['sentiment']=='positive']['review']
bad_reviews=train_df[train_df['sentiment']=='negative']['review']

print(f"First 10 samples of good reviews: {good_reviews[:10]}")
print(f"First 10 samples of bad reviews: {bad_reviews[:10]}")
```

```python
count=train_df['sentiment'].value_counts()
print('Total Counts of both sets',count)

print("=============")

#create a function to plot the counts using matplotlib
import matplotlib.pyplot as plt

def plot_counts(count_good,count_bad):
    labels = ['Positive Reviews', 'Negative Reviews']
    sizes = [count_good, count_bad]
    colors = ['lightblue', 'lightcoral']
    explode = (0.1, 0)  # explode the 1st slice

    plt.pie(sizes, explode=explode, labels=labels, colors=colors,
            autopct='%1.1f%%', shadow=True, startangle=140)

    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.title('Distribution of Positive and Negative Reviews')
    plt.show()
```
